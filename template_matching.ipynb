```
clahe = cv2.createCLAHE(clipLimit=2.0, tileGridSize=(8,8))
img_gr = cv2.cvtColor(img,cv2.COLOR_RGB2GRAY)
img_clahe = clahe.apply(img_gr)
plt.imshow(img_clahe,cmap ="gray")
```

```

img_edge = cv2.Canny(img_gr,100,100)
plt.imshow(img_edge,cmap = 'gray')
```

```
img_temp_gr = cv2.cvtColor(img_temp,cv2.COLOR_RGB2GRAY)
h ,w = img_temp_gr.shape
img_temp_edge = cv2.Canny(img_temp_gr,100,100)
plt.imshow(img_temp_edge,cmap="gray")
```

```
img_temp_clahe = clahe.apply(img_temp_gr)
plt.imshow(img_temp_clahe,cmap = 'gray')
```

```
img_temp_clahe_edge = cv2.Canny(img_temp_clahe,100,100)
plt.imshow(img_temp_clahe_edge,cmap="gray")
```

```
result = cv2.matchTemplate(img_clahe,img_temp_clahe,cv2.TM_CCOEFF_NORMED)
min_value , max_value, min_loc, max_loc = cv2.minMaxLoc(result)
top_left = max_loc
bottom_right = (top_left[0]+w,top_left[1]+h)
img_result = img.copy()
cv2.rectangle(img_result, top_left, bottom_right,255,2)
plt.imshow(img_result)
```

```
cv2.imwrite("/content/drive/MyDrive/work/pair0010_frm1_Coke_0217_result.jpg",cv2.cvtColor(img_result, cv2.COLOR_RGB2BGR))
```
